```python
# work without spacing and regex 

a={"how":"I'm Good","where":"I'm from Pakistan","goodbye":"Good bye"}

stopwords=['is','am','are','the','a','am','of','from','in','on','you']

punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

print("Bot: Hi")
while True:
    
    #take input and make lower case:
    b=input("Human: ").lower()
   
    # Remove punctuation:
    punct = ""       
    for char in b:
        if char not in punctuations:
            punct = punct + char
   
   
    # remove stop words (create a b_list of b(human quest) then remove stop words using if else)
    b1 = ""
    b_list=punct.split(" ")

    for i in b_list:
        if i in stopwords:
            pass
        else:
            b1+=i
          
        
    # remove space   (irrelevant)
    b=b1.replace(" ","")
    
    
    # search for answer from dictionary(a)
    if b in a.keys():
        print("\t Bot: ",a[b])
        
        # check for conversation end
        if (a[b]=="Good bye"):
            break
    else:
        print("\tBot: irrelevant question")
```

```python
import re
import json
from nltk.corpus import stopwords

    
with open("ques_ans.json","r") as js:
    questions=json.load(js)
    
sw=stopwords.words('english')


print("Bot: Hi")

while True:
    #take input and make lower case:
    sen=input("Human: ").lower()
   
    # Remove punctuation:
    sen = re.sub(r'[^\w\s]','',sen)
    
    # remove stop words
    sen1 = ""
    b_list=sen.split(" ")

    for i in b_list:
        if i in sw:
            pass
        else:
            sen1+=i
            sen1+=" "
    
    # remove space 
    sen1=sen1.strip()
    
    # search question in file 
    if sen1 in questions.keys():
        print("\t Bot: ",questions[sen1])
        
        if (questions[sen1]=="Good bye"):
            break
    
    # check for irrelevant question
    else:
        print("\tBot: Irrelevant question")
```
