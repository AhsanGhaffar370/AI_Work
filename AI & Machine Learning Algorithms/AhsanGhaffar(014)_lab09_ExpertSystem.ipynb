```python
print("***********Find Animal***********")
print("\nInstructions:")
print("For every question press y for *YES* and any other letter for *NO*")

def take_in(quest):
    quest1=("\n\nQ: The Animal "+quest+"\nAns: ")
    
    ans= input(quest1).lower()
    
    if(ans=="y"):
        return True
    else:
        return False

    
def find_animal():

      ############################
     ## Questions for mammals  ##
    ############################
    hair = take_in("has hair?")
    if (hair):
        milk = take_in("gives milk?")
        if (milk): # If true then it is MAMMAL
            
            meat = take_in("eats meat?")
            if(meat): # If true then it is CARNIVORE
                tawny = take_in("has a tawny colour?")
                if(tawny):
                    spots = take_in("has dark spots?")
                    if(spots):
                        return "cheetah"

                    stripes = take_in("has black stripes?")
                    if(stripes):
                        return "tiger"


            teeth = take_in("has pointed teeth?")
            if(teeth):
                claws = take_in("has claws?")
                if(claws):
                    eyes = take_in("'s eyes point forward?")
                    if(eyes): # If true then it is CARNIVORE
                        tawny = take_in("has a tawny colour?")
                        if(tawny):
                            spots = take_in("has dark spots?")
                            if(spots):
                                return "cheetah"

                            stripes = take_in("has black stripes?")
                            if(stripes):
                                return "tiger"


            hooves = take_in("has hooves?")
            if(hooves): # If true then it is UNGULATE
                legs = take_in("has long legs?")
                if(legs):
                    neck= take_in("has long neck?")
                    if(neck=="y"):
                        spots= take_in("has dark spots?")
                        if(spots):
                            return "giraffe"
                        
                color= take_in("has a white color?")
                if(color):
                    stripes= take_in("has black stripes?")
                    if(stripes):
                        return "zebra"

            chews = take_in("chews cud?")
            if(chews): # If true then it is UNGULATE AND EVEN-TOED
                return "ungulate and even-toed"

            
      ############################
     ### Questions for birds  ###
    ############################
    feathers= take_in("has feathers?")
    if(feathers): # If true then it is BIRD
        no_fly= take_in("does not fly?")
        if(no_fly):
            legs= take_in("has long legs?")
            if(legs):
                neck= take_in("has a long neck?")
                if(neck):
                    black= take_in("is black and white?")
                    if(black=="y"):
                        return "ostrich"


            swim= take_in("swims?")
            if(swim):
                black= take_in("is black and white?")
                if(black):
                    return "penguin"

        flier= take_in("is good flier?")
        if(flier):
            return "albatross"    


    flies= take_in("flies?")
    if(flies):
        eggs= take_in("lays eggs?")
        if(eggs): # If true then it is BIRD
            no_fly= take_in("does not fly?")
            if(no_fly):
                legs= take_in("has long legs?")
                if(legs):
                    neck= take_in("has a long neck?")
                    if(neck):
                        black= take_in("is black and white?")
                        if(black=="y"):
                            return "ostrich"


                swim= take_in("swims?")
                if(swim):
                    black= take_in("is black and white?")
                    if(black):
                        return "penguin"

            flier= take_in("is good flier?")
            if(flier):
                return "albatross"    
        
    
animal = find_animal()

print("\n************************")
print("The animal is ",animal)
print("************************")
```
