```python
import numpy as np
import pandas as pd
import random

data_set=pd.read_csv("C:\\Users\\HP\\Desktop\\Git_AI_Work\\AI & Machine Learning Algorithms\\Iris.csv")
data_set.head(150)

data_set=data_set.sample(frac=1)
# data_set.head(150)
```

```python
features = data_set.values[:, 1:5]  #x
labels= data_set.values[: , 5]  # y
```

```python
#75% training and 25% testing
test=0
train=input("enter training percentage: ")
# test=str(100-int(test))
#find 75% of 149 records
```

```python
max_index=int(len(data_set)*int(train)/100)
# test1=len(data_set)-train1

feature_train=features[0:max_index,:]

feature_test=features[max_index:len(data_set),:]

label_train=labels[0:max_index]

label_test=labels[max_index:len(data_set)]
```

```python
print(features)
print(labels)
print(feature_train)
print(feature_test)
print(label_train)
print(label_test)
```

```python
from sklearn.neural_network import MLPClassifier

# clf = MLPClassifier( alpha=1e-5,hidden_layer_sizes=(2,1), random_state=1)
clf = MLPClassifier( alpha=(1),hidden_layer_sizes=(1,1), random_state=1)

clf.fit(feature_train, label_train)
# MLPClassifier(alpha=1e-05, hidden_layer_sizes=(1, 3), random_state=1,solver='lbfgs')
```

```python
label_pred= clf.predict(feature_test)
```

```python
count=0
for t,p in zip(label_test, label_pred):
    print(t+"->"+p)
    count=count+1
print("Total tsetings:",count)

count1=0
for t,p in zip(label_test, label_pred):
    if(t==p):
        count1=count1+1
    
print("correct pred:",count1)
acc=(count1/len(label_test))*100

print("Accuracy",acc)
```
