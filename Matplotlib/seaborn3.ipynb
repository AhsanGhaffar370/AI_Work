# Plotting with Categorical Data:
### This approach comes into the picture when our main variable is further divided into discrete groups  (categorical). We use catplot to analyze non-numeric data.

## catplot():

### 1. Scatterplots – These include stripplot(), swarmplot()

### 2. Distribution Plots – which are boxplot(), violinplot(), boxenplot()

### 3. Estimateplots – namely pointplot(), barplot(), countplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)

a = sns.load_dataset("tips")

sns.catplot(x="day", y="total_bill", data=a) # by default we have kind=scatter
```

```python
## Count plot
df=sns.load_dataset("tips")

sns.catplot('sex',data=df, kind='count') # by default we have x-axis
```

```python
sns.catplot(y='sex',data=df, kind='count')
```

```python
sns.catplot(x='sex', y='tip',data=df, kind='bar')
```

```python
a = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", kind="box", data=a) # by default we have kind=scatter
# the dots above boxes are outliers
```

```python
a = sns.load_dataset("tips")
sns.catplot(kind="box", data=a, orient='v') # show all features on x-axis and show percentage on y-axis
# the dots above boxes are outliers
```

```python
a = sns.load_dataset("tips")
sns.catplot(x="day", kind="count", data=a) # by default we have kind=scatter
```

```python
a = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", kind="boxen", data=a) # by default we have kind=scatter
```

# Visualizing the distribution of a dataset:
#### This basically deals with understanding datasets with context to being univariate or bivariate. 
### we use dist or joint plot for numeric data analysis

## Plotting Univariate distributions (distplot):

#### When we have to work with only one independent variables/features, we use univariant plotting distribution. The best function to plot these type of graphs is distplot().

```python
df=sns.load_dataset("tips")

sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Plotting bivariate distributions (jointplot):

#### When we have to work with two independent variables/features, we use bivariant plotting distribution. The best function to plot these type of graphs is jointplot().

```python
x=pd.DataFrame({'Day':[1,2,3,4,5,6,7],
                'Grocery':[30,80,45,23,51,46,76],
                'Clothes':[13,40,34,23,54,67,98],
                'Utensils':[12,32,27,56,87,54,34]}, index=[1,2,3,4,5,6,7])

y=pd.DataFrame({'Day':[8,9,10,11,12,13,14],
                'Grocery':[30,80,45,23,51,46,76],
                'Clothes':[13,40,34,23,54,67,98],
                'Utensils':[12,32,27,56,87,54,34]},index=[8,9,10,11,12,13,14])

mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)

with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="kde", color="b")
```

```python
df=sns.load_dataset("tips")

sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Multi-Plot Grids:
#### Python Seaborn allows you to plot multiple grids side-by-side. These are basically plots or graphs that are plotted using the same scale and axes to aid comparison between them. This, in turn, helps the programmer to differentiate quickly between the plots and obtain large amounts of information.

### FACEGRID()

```python
sns.set(style="darkgrid")

a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col="species")

b.map(plt.hist, "sepal_length")
```

### PAIRGRID()

```python
sns.set(style="ticks")

a = sns.load_dataset("flights")
b = sns.PairGrid(a)

b.map(plt.scatter)
```
