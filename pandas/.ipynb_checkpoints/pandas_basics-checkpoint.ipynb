```python
#how to reindex and use of fill value of paramenter method

import pandas as pd
obj3 = pd.Series(['blue','purple','yellow',], index= [0,3,6])
print(obj3)
```

## Things to remainder with every function
### inplace= True or False --> If True, do operation on original data. If false, do operation on copy of data and return it. By default it's False.
### axis=0 or 1 --> Axis to target row or column. Can be either the axis name (‘index’, ‘columns’) or number (0, 1).

### df.loc["index"] or df.loc[row, column] -->have to use proper row or col name
### df.iloc["index"] or df.loc[row, column] --> have to use index num of row or column
### df.drop(index) -->delete row
### df.drop(col_name, axis=column) --> delete column
### df.drop(col_name, axis=column, inplace=True) --> delete column from original dataframe
### df.sort_index()
### df.sort_value()
### df.sort_value(by='col_name') or by=['col1', 'col2', .. ]

### df.dropna()
### df.dropna(how="all")  or how= 'any'
### df.dropna(axis=1) --> to drop column contain nan 
### df.dropna(thresh=1) --> thresh=1 means, row contain atleast one valid/non NAN value than it's not gonna drop
### df.drop_duplicates() 
### df.drop_duplicates(keep="first") --> only take first row and remove all duplicates of it 
### df.drop_duplicates(keep="last") --> only take last row and remove all duplicates of it 
### df.drop_duplicates(keep=False)  --> remove all duplicate rows
### df.drop_duplicates(subset=[col_name]) --> drop duplicate wrt specific col
### df.fillna(value)
### df.interpolate() --> fill nan values with some meaningful info
### df.info
### df.shape
### pd.merge(df1, df2)

```python
state = ['Ohio','Ohio','Ohio','Navada','Navada','Navada',]

data= {'state':state,
       'year': [2000,2002,2001,2002,2003, 2005],
       'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}


frame3 = pd.DataFrame(data, columns=['year', 'state', 'pop'],index=['one','two','three','four','five','six'])

frame3["dept"]=["CS","EE","MA","IT","SE","BBA"]

frame3.head()
```

```python
frame3.loc['seven']={'year':2003, 'state':'Navada', 'pop':2.9, 'dept':'SE'}
```

```python
frame3['avg']=np.nan #-->create new column and fill it with nan values
```

```python
new=pd.Series([55,23,77],index=['one','three','five'])
```

```python
frame3['test']=new
```

```python
frame3
```

```python
frame3.dropna()
```

```python
frame3.dropna(axis='columns')
```

```python
# Drop the rows or column where all elements are missing.
frame3.dropna(how='all')
```

```python

frame3.dropna(how='all', axis=1)
```

```python
# Drop the rows or column where any one elements is missing.
frame3.dropna(how='any', axis=1)
```

```python
frame3.dropna(thresh=5) # it means a row should contain atleast 5 non NAN values
```

```python
frame3.loc['seven','test']=77.0
frame3
```

```python
frame3.drop_duplicates()
```

```python
frame3.drop_duplicates(subset=['state']) ##drop duplicates with respect to state column
```

```python
frame3.drop_duplicates(subset=['state'], keep='last')
```

```python
frame3.head()
```

```python
frame3.fillna({'test':0.5})
```

```python
values = {'avg': 0, 'state': 1, 'pop': 2, 'dept': 3}
frame3.fillna(value=values)
```

```python
frame3
```

```python
## only fill test col values because it have some missing values so it fills nan values by finding average of 
# upper and lower value of nan value 
# suppose we have 55 - NAN - 23 so we fill 39 inplace of NAN
frame3.interpolate() 
```

# dataframe groups

### df.groupby('col_name')
### df.get_group('col element name')
### df.groupby('col_name').agg([fun1, func2, func3, .. ])
### pd.concat([df1,df2] , ignore_index=True)
### pd.merge(df2, df1, on='key', how="outer") or how="inner" is by default or we can use "left"/"right"
### df.stack() -->convert most inner header into row
### df.stack(limit=0) --> convert most outer header into row
### df.unstack()

```python
g=frame2.groupby('state')
```

```python
g.get_group('Navada')
```

```python
g.max()
```

```python
g.mean()
```

```python
g.describe()
```

```python
%matplotlib inline
g.plot()
```

```python
#if we want to add required columns in group 

g=pd.DataFrame(frame2.groupby('state')['pop'].mean())
g
```

```python
frame2.groupby('state').pop.mean()
```

```python
frame2.groupby('state').pop.agg(['count','mean', 'min']) # apply multiple function at once 
```

```python
g['pop_count']=pd.DataFrame(frame2.groupby('state')['pop'].count())
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],'data2': range(3)})
```

```python
pd.concat([df1,df2])
```

```python
conc= pd.concat([df1,df2] , ignore_index=True)
conc
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
## it's an example of inner join where it takes common values from both dataframe
## inner join is default but we can set outer using how="inner" or outer

pd.merge(df2, df1, on='key')
```

```python
pd.merge(df2, df1, on='key', how="outer")
```
